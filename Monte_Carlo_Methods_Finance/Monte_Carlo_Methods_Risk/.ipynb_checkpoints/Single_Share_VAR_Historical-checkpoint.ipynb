```python
# VAR calcuations for a ingle share using standard VAR and historical VAR to compare the estimates
```

```python
#import modules
import numpy as np
from scipy.stats import uniform
from scipy.stats import norm
import math
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
# Market, modelling, share dynamics and information 
S_0   = 25
sigma = 0.1
mu    = 0.1
T     = 1
alpha = 0.01
dT    = 1/365
```

```python
# Share path function 
# Function to generate synthetic share prices over time until final period
def share_path(S_0,r,sigma,Z,dT):
    """
    Generates share prices given some random normal values (Z) 
    with adjustment of cumsum to capture share prices over time as 
    result of individual changes up to that point
    """
    return S_0*np.exp(np.cumsum((r-sigma**2/2)*dT + sigma*np.sqrt(dT)*Z,1))
```

```python
# Generation of historical data 
Z_hist       = norm.rvs(size = [1,10* int(1/dT)])
price_path   = share_path(S_0,mu,sigma,Z_hist,dT)

hist_returns = np.log(price_path[0,1:])-np.log(price_path[0,0:-1])
# returns above calculed np.log(next value/previous ) = np.log(next)-np.log(prev)
mu_est   = np.mean(hist_returns)*365
sigma_est = np.std(hist_returns)*np.sqrt(365)
```

```python
# Function for Building historical prices into future to get set of returns; assumes Geometric Brownian motion
def adj_terminal_share_price(S_0,mu_est,sigma_est,Z,T):
    return S_0* np.exp(mu_est*T + sigma_est*np.sqrt(T)*Z)
```

```python
# Calculation standard VAR_estimate for different sizes of N where N =1000,2000...50,000

curr_share_price = price_path[0][-1]
curr_port_value =  curr_share_price

VAR_estimates = [None]* 50

for i in range(1,51):
    Z = norm.rvs(size = i*1000)
    future_port_value =  adj_terminal_share_price(curr_share_price,mu_est,sigma_est,Z,T)
    port_returns      =  (future_port_value-curr_port_value) / curr_port_value
    port_returns      = np.sort(port_returns)
    VAR_estimates[i-1]  = -1*port_returns[int(np.floor(alpha*i*1000))-1] 
```

```python
# Calculation historical VAR estimate

Hist_estimates = [None]*50 # Historical VAR values storage

for i in range(1,51):
    
    port_returns = [None]*i*1000
    
    for j in range(1, i*1000+1):
        
        rand_samp = uniform.rvs(size = 365) * (len(price_path[0])-1)
        rand_samp = [int(x) for x in rand_samp]
        returns   = hist_returns[rand_samp]
        projected  = curr_share_price*np.exp(np.sum(returns,axis=0))
        Hist_estimates[j-1]  = (projected - curr_port_value)/curr_port_value
```

```python
# Plotting of values to compare the two 
plt.plot([i*1000 for i in range(50)],VAR_estimates)
plt.xlabel("Different sizes of n")
plt.ylabel("VAR value")
plt.show()
```

```python
# Plotting of values to compare the two 
plt.plot([i*1000 for i in range(50)],Hist_estimates)
plt.xlabel("Different sizes of n")
plt.ylabel("VAR value")
plt.show()
```
